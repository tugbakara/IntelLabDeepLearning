## **Data Augmentation**

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/home/tugbakara/YL/semester_II/dersler/deeplearning/augmented_sensor_data_combined.csv')
print("first",data.shape)
def inject_abnormal_data(data):
    new_data = pd.DataFrame(columns=data.columns)
    for i in range(len(data)):
        new_data = new_data.append(data.iloc[i], ignore_index=True)
        if (i + 1) % 4 == 0:
            temperature_noise = np.random.choice(np.concatenate((np.random.uniform(0, 10, len(data)), np.random.uniform(45, 125, len(data)))))
            humidity_noise = np.random.choice(np.concatenate((np.random.uniform(-1, 0, len(data)), np.random.uniform(100, 150, len(data)))))
            voltage_noise = np.random.choice(np.concatenate((np.random.uniform(0, 1.99, len(data)), np.random.uniform(3.1, 4, len(data)))))
            light_noise = np.random.uniform(0, 400, len(data))

            new_row = {
                'temperature': temperature_noise,
                'humidity': humidity_noise,
                'voltage': voltage_noise,
                'light': light_noise[i // 4],
                'temperature_anomaly': 1 if (temperature_noise < 10 or temperature_noise > 45) else 0,
                'humidity_anomaly': 1 if (humidity_noise < 0 or humidity_noise > 100) else 0,
                'voltage_anomaly': 1 if (voltage_noise < 2 or voltage_noise > 3) else 0,
                'light_anomaly': 1 if light_noise[i // 4] < 400 else 0
            }
            new_data = new_data.append(new_row, ignore_index=True)

    return new_data

data = inject_abnormal_data(data)
data.to_csv('/home/tugbakara/YL/semester_II/dersler/deeplearning/augmented_sensor_data_combined2.csv', index=False)
print("last",data.shape)
```

## **1D Preprocessing**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd


data = pd.read_csv('/content/drive/MyDrive/deeplearning/deeplearning_final/datas/labeled_sensor.csv')
nan_values = data.isna().sum()
print(data.head)
columns_with_nan = nan_values[nan_values > 0].index.tolist()
if len(columns_with_nan) > 0:
    print("NaN değerleri olan sütunlar:")
    print(columns_with_nan)
else:
    print("Veri setinde NaN değeri olan sütun bulunmamaktadır.")
anomaly_columns = ['temperature_anomaly', 'voltage_anomaly', 'light_anomaly', 'humidity_anomaly']
data[anomaly_columns] = data[anomaly_columns].astype(int)

unique_moteids = data['moteid'].unique()
print("Unique Mote IDs:", unique_moteids)

data = data.drop(columns=['date','epoch','moteid'])

X = data.drop(columns=["temperature_anomaly", "voltage_anomaly", "humidity_anomaly","light_anomaly"])
y = data[["temperature_anomaly", "voltage_anomaly", "humidity_anomaly","light_anomaly"]]
print("*******")
print(X.head)
print("-------")
print(y.head)
scaler = MinMaxScaler()
normalized_columns = ['temperature', 'humidity', 'voltage','light']
X_normalized = scaler.fit_transform(X[normalized_columns])

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.01, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.01, random_state=42)
```

```python
import numpy as np
print("x_train:", X_train.shape)
print("y_train:", y_train.shape)
print("x_test:", X_test.shape)
print("y_test:", y_test.shape)
print("x_val", X_val.shape)
print("y_val:", y_val.shape)
print(X_train)
```

## **Models**

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import History
from sklearn.utils import class_weight

input_shape = (4, 1)
input_layer = Input(shape=input_shape)

conv0 = Conv1D(filters=128, kernel_size=1, activation='relu')(input_layer)
pool0 = MaxPooling1D(pool_size=1)(conv0)
conv1 = Conv1D(filters=64, kernel_size=1, activation='relu')(pool0)
pool1 = MaxPooling1D(pool_size=1)(conv1)
conv2 = Conv1D(filters=32, kernel_size=1, activation='relu')(pool1)
pool2 = MaxPooling1D(pool_size=1)(conv2)
flattened = Flatten()(pool2)
dense1 = Dense(64, activation='relu')(flattened)
dense2 = Dense(32, activation='relu')(dense1)
dense3 = Dense(16, activation='relu')(dense2)
temp_output = Dense(2, activation='softmax', name='temperature_output')(dense3)
volt_output = Dense(2, activation='softmax', name='voltage_output')(dense3)
humid_output = Dense(2, activation='softmax', name='humidity_output')(dense3)
light_output = Dense(2, activation='softmax', name='light_output')(dense3)

model = Model(inputs=input_layer, outputs=[temp_output, volt_output, humid_output, light_output])

model.compile(optimizer=Adam(learning_rate=0.001),
              loss=['sparse_categorical_crossentropy'] * 4,
              metrics=['accuracy'])

history = model.fit(X_train, [y_train['temperature_anomaly'], y_train['voltage_anomaly'],
                              y_train['humidity_anomaly'], y_train['light_anomaly']],
                    validation_data=(X_val, [y_val['temperature_anomaly'], y_val['voltage_anomaly'],
                                             y_val['humidity_anomaly'], y_val['light_anomaly']]),
                    batch_size=32, epochs=16, verbose=1)

print("Validation Accuracy:")
print("Temperature:", history.history['val_temperature_output_accuracy'])
print("Voltage:", history.history['val_voltage_output_accuracy'])
print("Humidity:", history.history['val_humidity_output_accuracy'])
print("Light:", history.history['val_light_output_accuracy'])

print("Validation Loss:")
print("Temperature:", history.history['val_temperature_output_loss'])
print("Voltage:", history.history['val_voltage_output_loss'])
print("Humidity:", history.history['val_humidity_output_loss'])
print("Light:", history.history['val_light_output_loss'])

model.save("/content/drive/MyDrive/deeplearning/deeplearning_final/models/2.h5")
```

```python
from tensorflow.keras.models import load_model
from sklearn.metrics import confusion_matrix, accuracy_score, precision_recall_fscore_support
import numpy as np

model = load_model("/content/drive/MyDrive/deeplearning/deeplearning_final/models/2.h5")

y_pred = model.predict(X_test)

class_mapping = {
    'temperature_0': 0, 'temperature_1': 1,
    'voltage_0': 0, 'voltage_1': 1,
    'humidity_0': 0, 'humidity_1': 1,
    'light_0': 0, 'light_1': 1
}

metrics = {
    'temperature': {},
    'voltage': {},
    'humidity': {},
    'light': {}
}

overall_true = []
overall_pred = []

for sensor_type, idx in class_mapping.items():
    if 'temperature' in sensor_type:
        y_true = y_test['temperature_anomaly'].map(lambda x: 'No Anomaly' if x == 0 else 'Anomaly')
        y_pred_sensor = np.argmax(y_pred[0], axis=1)
    elif 'voltage' in sensor_type:
        y_true = y_test['voltage_anomaly'].map(lambda x: 'No Anomaly' if x == 0 else 'Anomaly')
        y_pred_sensor = np.argmax(y_pred[1], axis=1)
    elif 'humidity' in sensor_type:
        y_true = y_test['humidity_anomaly'].map(lambda x: 'No Anomaly' if x == 0 else 'Anomaly')
        y_pred_sensor = np.argmax(y_pred[2], axis=1)
    elif 'light' in sensor_type:
        y_true = y_test['light_anomaly'].map(lambda x: 'No Anomaly' if x == 0 else 'Anomaly')
        y_pred_sensor = np.argmax(y_pred[3], axis=1)

    overall_true.extend(y_true)
    overall_pred.extend(np.where(y_pred_sensor == 0, 'No Anomaly', 'Anomaly'))

    cm = confusion_matrix(y_true, np.where(y_pred_sensor == 0, 'No Anomaly', 'Anomaly'))

    labels = ['No Anomaly', 'Anomaly']
    metrics[sensor_type.split('_')[0]] = {}
    for i, label in enumerate(labels):
        accuracy = accuracy_score(y_true, np.where(y_pred_sensor == i, label, 'Anomaly'))
        precision, recall, f1_score, _ = precision_recall_fscore_support(y_true, np.where(y_pred_sensor == i, label, 'Anomaly'), average='weighted')

        metrics[sensor_type.split('_')[0]][label] = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1_score
        }

for sensor_type, label_metrics in metrics.items():
    print(f"{sensor_type.capitalize()} Sensor Metrics:")
    for label, metric in label_metrics.items():
        print(f"Label: {label}")
        print(f"Accuracy: {metric['accuracy']:.4f}")
        print(f"Precision: {metric['precision']:.4f}")
        print(f"Recall: {metric['recall']:.4f}")
        print(f"F1-score: {metric['f1_score']:.4f}")
        print()
    print()

overall_accuracy = accuracy_score(overall_true, overall_pred)
overall_precision, overall_recall, overall_f1_score, _ = precision_recall_fscore_support(overall_true, overall_pred, average='weighted')

print("Overall Metrics:")
print(f"Overall Accuracy: {overall_accuracy:.4f}")
print(f"Overall Precision: {overall_precision:.4f}")
print(f"Overall Recall: {overall_recall:.4f}")
print(f"Overall F1-score: {overall_f1_score:.4f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten, Dropout, Reshape, BatchNormalization, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(32, 32, 1)))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(64, (1, 1), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(256, (1, 1), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32 * 32 * 8, activation='softmax'))
model.add(Reshape((32, 32, 8)))
optimizer = Adam(learning_rate=0.0001)
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, batch_size=8, epochs=32, validation_data=(X_val, y_val))
model.save("/content/drive/MyDrive/deeplearning/deeplearning_final/models/2d_cnn_model_8class_pooling1.h5")
model.summary()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Conv2D, Dense, Dropout, Reshape, BatchNormalization, GlobalAveragePooling2D, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import MobileNetV2, ResNet50V2, ResNet101V2, DenseNet121, EfficientNetV2B0
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import img_to_array, array_to_img
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

def create_transfer_model(base_model, input_shape, num_classes):
    base_model.trainable = False

    inputs = Input(shape=input_shape)
    x = Conv2D(3, (3, 3), padding='same', trainable=False)(inputs)
    x = base_model(x, training=False)
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = Dropout(0.5)(x)
    outputs = Dense(input_shape[0] * input_shape[1] * num_classes, activation='softmax')(x)
    outputs = Reshape((input_shape[0], input_shape[1], num_classes))(outputs)

    model = Model(inputs, outputs)
    return model

input_shape = (32, 32, 1)
num_classes = 8

base_models = {
    'MobileNetV2': MobileNetV2(include_top=False, weights='imagenet', input_shape=(32, 32, 3)),
    'ResNet50V2': ResNet50V2(include_top=False, weights='imagenet', input_shape=(32, 32, 3)),
    'ResNet101V2': ResNet101V2(include_top=False, weights='imagenet', input_shape=(32, 32, 3)),
    'DenseNet121': DenseNet121(include_top=False, weights='imagenet', input_shape=(32, 32, 3)),
    'EfficientNetV2B0': EfficientNetV2B0(include_top=False, weights='imagenet', input_shape=(32, 32, 3))
}


results = {}
saved_models = {}

for model_name, base_model in base_models.items():
    print(f'Training {model_name}...')
    model = create_transfer_model(base_model, input_shape, num_classes)
    model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model.summary()
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)

    history = model.fit(X_train, y_train, batch_size=8, epochs=8, validation_data=(X_val, y_val), callbacks=[reduce_lr])

    save_path = f"/content/drive/MyDrive/deeplearning/deeplearning_final/models/{model_name}_transfer_learning1.h5"
    model.save(save_path)
    saved_models[model_name] = save_path

    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    results[model_name] = accuracy
    print(f'{model_name} Test accuracy: {accuracy:.4f}')

```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import History

input_shape = (4, 1)
input_layer = Input(shape=input_shape)

conv1 = Conv1D(filters=256, kernel_size=1, activation='relu')(input_layer)
batch1 = BatchNormalization()(conv1)
pool1 = MaxPooling1D(pool_size=1)(batch1)
drop1 = Dropout(0.5)(pool1)
conv2 = Conv1D(filters=128, kernel_size=1, activation='relu')(drop1)
batch2 = BatchNormalization()(conv2)
pool2 = MaxPooling1D(pool_size=1)(batch2)
drop2 = Dropout(0.5)(pool2)
conv3 = Conv1D(filters=64, kernel_size=1, activation='relu')(drop2)
batch3 = BatchNormalization()(conv3)
pool3 = MaxPooling1D(pool_size=1)(batch3)
drop3 = Dropout(0.5)(pool3)
conv4 = Conv1D(filters=32, kernel_size=1, activation='relu')(drop3)
batch4 = BatchNormalization()(conv4)
pool4 = MaxPooling1D(pool_size=1)(batch4)
drop4 = Dropout(0.5)(pool4)
flattened = Flatten()(drop4)
dense1 = Dense(64, activation='relu')(flattened)
batch5 = BatchNormalization()(dense1)
drop5 = Dropout(0.5)(batch5)
dense2 = Dense(32, activation='relu')(drop5)
batch6 = BatchNormalization()(dense2)
drop6 = Dropout(0.5)(batch6)
dense3 = Dense(16, activation='relu')(drop6)
batch7 = BatchNormalization()(dense3)
drop7 = Dropout(0.5)(batch7)

temp_output = Dense(2, activation='softmax', name='temperature_output')(drop7)
volt_output = Dense(2, activation='softmax', name='voltage_output')(drop7)
humid_output = Dense(2, activation='softmax', name='humidity_output')(drop7)
light_output = Dense(2, activation='softmax', name='light_output')(drop7)

model = Model(inputs=input_layer, outputs=[temp_output, volt_output, humid_output, light_output])

model.compile(optimizer=Adam(learning_rate=0.0001),
              loss=['sparse_categorical_crossentropy'] * 4,
              metrics=['accuracy'])

history = model.fit(X_train, [y_train['temperature_anomaly'], y_train['voltage_anomaly'],
                              y_train['humidity_anomaly'], y_train['light_anomaly']],
                    validation_data=(X_val, [y_val['temperature_anomaly'], y_val['voltage_anomaly'],
                                             y_val['humidity_anomaly'], y_val['light_anomaly']]),
                    batch_size=32, epochs=16, verbose=1)

print("Validation Accuracy:")
print("Temperature:", history.history['val_temperature_output_accuracy'])
print("Voltage:", history.history['val_voltage_output_accuracy'])
print("Humidity:", history.history['val_humidity_output_accuracy'])
print("Light:", history.history['val_light_output_accuracy'])

print("Validation Loss:")
print("Temperature:", history.history['val_temperature_output_loss'])
print("Voltage:", history.history['val_voltage_output_loss'])
print("Humidity:", history.history['val_humidity_output_loss'])
print("Light:", history.history['val_light_output_loss'])

model.save("/content/drive/MyDrive/deeplearning/deeplearning_final/models/2_augmented.h5")
```
